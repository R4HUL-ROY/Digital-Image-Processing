#Task 1

```python
'''  
TASK - 1
Write a program to get the adaptive threshold to convert any grey image to the corresponding
black and white image.
'''

import cv2
import numpy as np
from google.colab.patches import cv2_imshow


def convert_to_black_and_white(img):
    row, col = img.shape
    threshold = (np.max(img) + np.min(img)) / 2
    for i in range(row):
        for j in range(col):
            if img[i,j] <= threshold:
                img[i,j] = 0
            else:
                img[i,j] = 255
    return img



# Read the Image
path = r'/content/3.jpg'
img = cv2.imread(path, 0)
# original image
cv2_imshow(img)
convert_to_black_and_white(img)
# After performing operation black and white image
cv2_imshow(img)
```

#Task 2

```python
 '''
    TASK - 2
    Make your own method to implement contrast streching in a given image.
 '''


import cv2
import numpy as np
from google.colab.patches import cv2_imshow


def contrast_strech(img):
    row, col = img.shape
    min = np.min(img)
    max = np.max(img)
    for i in range(row):
        for j in range(col):
            img[i,j] = ((img[i,j] - min) * (255/(max-min)))

    return img

# Read the Image
path = r'/content/8.jpg'
img = cv2.imread(path, 0)
# original image
cv2_imshow(img)
contrast_strech(img)
# After performing contrast streching
cv2_imshow(img)
```

```python
'''
    TASK - 3
    Remove the unnecessary area of an image using adaptive grey level slicing. The area contains
    lower intensity value is called unnecessary area.
'''

import cv2
import numpy as np
from google.colab.patches import cv2_imshow


def grey_slicing(img, x, y):
    x,y = int(x), int(y)
    row, col = img.shape
    for i in range(row):
        for j in range(col):
            if img[i,j] < x or img[i,j] > y:
                img[i,j] = 0
    return img

# Read the Image
path = r'/content/1..jpg'
img = cv2.imread(path, 0)
x, y = input("Enter slicing Range : ").split()
x, y = int(x), int(y)
# original image
cv2_imshow(img)
grey_slicing(img, x, y)
# After performing grey-scale slicing
cv2_imshow(img)
```
