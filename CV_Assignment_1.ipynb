#Task1 : 
**Compute the standard deviation, mean, mode and median of an image**

```python
'''Compute the standard deviation, mean, mode and median of an image'''

import numpy as np
from statistics import mode
print(f"the mean is : {np.mean(img)}")
print(f"the median is : {np.median(img)}")
print(f"standard deviation of the image is : {np.std(img)}")
print(f"mode of the array : {mode(list(img.flatten()))}")
```

#Task2 : 
**Compute the frequency of each intensity level from an image**

```python
'''Compute the frequency of each intensity level from an image'''

print(img)
flatten_arr = list(img.flatten())
freq_dict = {}
for i in range(0,256):
    freq_dict[i] = flatten_arr.count(i)
print(freq_dict)
```

**Intensity slicing**

```python
## intensity slicing

temp_img = img.copy()
print(temp_img)
row, col = temp_img.shape

n1, n2 = input("enter filter range(0-255) : ").split()
n1, n2 = int(n1), int(n2)
for i in range(row):
    for j in range(col):
        if temp_img[i,j] < n1 or temp_img[i,j] > n2:
            temp_img[i,j] = 0
            
cv2_imshow(temp_img)
```

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

img = cv2.imread(r'/content/drive/MyDrive/computer vision/1..jpg', 0)
cv2_imshow(img)
```

```python
# frequency of each intensity pixel from an image, using counting sort technique

arr = np.zeros((256))
row, col = img.shape
for i in range(row):
    for j in range(col):
        arr[img[i,j]] += 1

print(arr)        
```

```python
import matplotlib.pyplot as plt
plt.plot(arr)
plt.show()
```
